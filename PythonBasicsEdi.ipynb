**Identity operator**
* It tells about the value whether they share the same reference address or not

```python
x = [1,2,3]
y = [1,2,3]
z = x
print(x is y) # although values are same but they both share different memory
print(x is z) # z is holding the reference of x which holds the same [1,2,3] value
```

##**Conditional Statements**
If, Else, Else IF
* IF - One condition
* Else - If all condtionsa re not satisfied
* elif - If you want to have multiple conditions

```python
age = 5
if age>=20:
  print('You are an adult')
elif age >= 13 <=18:
  print('You are a teenager')
else:
  print('You are a minor')
```

**Nested IF**

```python
marks = int(input('Enter your Mark'))
if marks >= 33:
  print('Passed')
  if marks >=85 and marks<=100:
    print('Grade A')
  elif marks>=60 and marks<85:
    print('Grade B')
  else:
    print('Grade C')
else:
  print('Failed')
```

###**Looping**
To repeat a tast for a given range, Sequence, or a condition
* In python for Loop, While Loop

**For Loop**

In for loop There is no chance of getting Infinte loop

```python
for i in range(1,11):
  print(i)
```

```python
list1 = ['Apple','Banana','Guava','Melon','Mango']
for i in list1:
  print(i)
```

**While loop**
- When we have Condition
- You have to be careful about getting into infinite loop

```python
x = 1
while x<=10:
  print(x)
  x +=1 # i = i+1
```

**Nested Loop**


```python
for i in range(2,7):
  for j in range(1,11):
    print(f'{i} x {j} = {i*j}')
  print()
```

```python
x = 10
y = 20
print('The value of x is',x,'the value of y is',y)
print(f'The value of x is {x} and the value of y is {y}') # format strings
```

```python
name = ['Anjana','Ram','Akshaya','Akansha']
values = [100,95,89,90]
# keyword alert : zip()
for i,j in zip(name,values):
  print(f'{i} score {j}')
```

**Lambda Function**
- One statement function

```python
#keyword : Lambda
# defined the func -> Lambda
add = lambda x,y : print(f'{x} + {y} = {x+y}')

# call the function
add(2,3)
```
